# Project 14: Find the Largest Number in a List

```python
# This project teaches how to find the largest number using loops and lists
```

```python
# Step 1: Ask the user to enter numbers separated by commas
numbers_input = input("Enter numbers separated by commas: ")
```

```python
# Step 2: Convert input string to a list of numbers
numbers_list = [float(x) for x in numbers_input.split(",")]
```

```python
# Step 3: Initialize the largest number with the first element
largest = numbers_list[0]
```

```python
# Step 4: Compare each number in the list
for num in numbers_list:
    if num > largest:
        largest = num
```

```python
# Step 5: Print the largest number
print(f"The largest number in the list is: {largest}")
```
